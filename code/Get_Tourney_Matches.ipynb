```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re 
```

```python
#Reading in player data and deleting useless columns

player_data = pd.read_csv('../data/TC3_Scoresheet_Scoresheet.csv')

player_data.drop(columns = ['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3'], inplace = True)

player_data = player_data.iloc[0:40]
player_data
```

```python
#Changed phili, inthecypher, and koto's name since their igns were incorrect


player_data.loc[34, ('Names')] = 'phi lii special'
player_data.loc[36, ('Names')] = 'koto vt'
player_data.loc[27, ('Names')] = 'in the cypher'

players = list(player_data['Names'])

players
```

```python
def get_tourney_matches(url):
    """Grabs the games from specified url. Keeps games from search criteria by day in this case the day of the tourney
    """
    
    games = []
    
    response = requests.get(url)
        
    soup = BeautifulSoup(response.text, 'lxml')
    
    games_div = soup.find_all('div', {'class' : re.compile(r'profile__match-history-v2__item placement*')})
    games_div.reverse()
    
    for game in games_div:
        time = game.find('div', {'class' : 'age'}).find('span')['title']
        if re.search(r'03-12-2023*', time):
            games.append(game)
            
    return games
```

```python
game_data = []

for player in players:
    player = player.replace(" ", "")
    url = "https://lolchess.gg/profile/na/" + player + "/s8/matches/normal"
    games = get_tourney_matches(url)
    game_data += games
    
print(len(game_data))
```

```python
def read_match_data(game):
    placement = int(game.find('div', {'class' : 'placement'}).text[1])
    level = int(game.find('span', {'class' : 'level'}).text)

    player_name = game.find('div', {'class' : 'participants'}).find_all('a')[placement - 1].text.strip().lower()
    
    traits = game.find_all('div', {'class' :re.compile(r'tft-hexagon-image*')})
    trait_list = []
    for trait in traits:
        trait_list.append(trait.find('img')['title'])
        
    augments = game.find('div', {'class':'augments'}).find_all('div', {'class':re.compile(r'augments*')})
    augment_list = []
    for augment in augments:
        augment_list.append([augment.find('img')['title'], len(augment['class']) > 1])

    units = game.find_all('div', {'class':'unit'})
    unit_list = []
    for unit in units:
        img = unit.find('img')
        star = img['src'][-5]
        cost = img['src'][-12]
        champ = unit.find('div', {'class':re.compile(r'tft-champion*')}).find('img')['alt']
        items = unit.find_all('img', {'class' : 'item'})
        item_list=[]
        for item in items:
            item_list.append(item['title'])

        unit_list.append([star, cost, champ, item_list])
    
    num_fons = len(unit_list) - level
        
    return([player_name, placement, level, trait_list, augment_list, unit_list, num_fons])
    # print(player_name, placement, level, trait_list, augment_list, unit_list, num_fons)
```

```python
read_match_data(game_data[0])
```

```python
match_data_list = []

for game in game_data:
    match_data_list.append(read_match_data(game))
    
match_data_list
    
match_df = pd.DataFrame(match_data_list, columns=['player', 'placement', 'level', 'traits', 'augments', 'units', "tactician's crowns"])
match_df
```

```python
match_df.to_csv('../data/match_data.csv', index = False)
```
